```python
### TEST FUNCTION: test_question3
import numpy as np
import scipy.io

def load_data(myID):
    try:
        # Load training and test data
        Numpyfile0 = scipy.io.loadmat(f'digit0_stu_train{myID}.mat')
        Numpyfile1 = scipy.io.loadmat(f'digit1_stu_train{myID}.mat')
        testfile0 = scipy.io.loadmat('digit0_testset.mat')
        testfile1 = scipy.io.loadmat('digit1_testset.mat')
        
        train0 = Numpyfile0.get('target_img')
        train1 = Numpyfile1.get('target_img')
        test0 = testfile0.get('target_img')
        test1 = testfile1.get('target_img')
        
        return train0, train1, test0, test1

    except FileNotFoundError as e:
        print(f"Error: {e}")
        return None, None, None, None
def predict_class(features, means_f1, vars_f1, means_f2, vars_f2):
    predictions = []
    
    for feature in features:
        # Feature values
        f1, f2 = feature
        
        # Calculate likelihoods for class 0
        likelihood_0_f1 = gaussian_likelihood(f1, means_f1[0], vars_f1[0])
        likelihood_0_f2 = gaussian_likelihood(f2, means_f2[0], vars_f2[0])
        likelihood_0 = likelihood_0_f1 * likelihood_0_f2
        
        # Calculate likelihoods for class 1
        likelihood_1_f1 = gaussian_likelihood(f1, means_f1[1], vars_f1[1])
        likelihood_1_f2 = gaussian_likelihood(f2, means_f2[1], vars_f2[1])
        likelihood_1 = likelihood_1_f1 * likelihood_1_f2
        
        # Predict the class with higher likelihood
        if likelihood_0 > likelihood_1:
            predictions.append(0)
        else:
            predictions.append(1)
    
    return np.array(predictions)


def extract_features(images):
    avg_brightness = np.mean(images, axis=(1, 2))
    std_brightness = np.std(images, axis=(1, 2))
    return avg_brightness, std_brightness

def gaussian_likelihood(x, mean, variance):
    coefficient = 1.0 / np.sqrt(2 * np.pi * variance)
    exponent = np.exp(-((x - mean) ** 2) / (2 * variance))
    return coefficient * exponent

def calculate_naive_bayes_parameters(images):
    avg_brightness, std_brightness = extract_features(images)
    mean_f1 = np.mean(avg_brightness)
    var_f1 = np.var(avg_brightness)
    mean_f2 = np.mean(std_brightness)
    var_f2 = np.var(std_brightness)
    return mean_f1, var_f1, mean_f2, var_f2

def main():
    myID = '7302'
    train0, train1, test0, test1 = load_data(myID)
    
    if train0 is None or train1 is None or test0 is None or test1 is None:
        return

    # Calculate Naive Bayes parameters for digit 0 and digit 1
    mean_f1_0, var_f1_0, mean_f2_0, var_f2_0 = calculate_naive_bayes_parameters(train0)
    mean_f1_1, var_f1_1, mean_f2_1, var_f2_1 = calculate_naive_bayes_parameters(train1)
    
    # Save Naive Bayes parameters to a file
    np.savez('naive_bayes_params.npz',
             mean_f1_0=mean_f1_0, var_f1_0=var_f1_0, mean_f2_0=mean_f2_0, var_f2_0=var_f2_0,
             mean_f1_1=mean_f1_1, var_f1_1=var_f1_1, mean_f2_1=mean_f2_1, var_f2_1=var_f2_1)
    
    # Extract features from test data
    avg_brightness_test0, std_brightness_test0 = extract_features(test0)
    avg_brightness_test1, std_brightness_test1 = extract_features(test1)
    
    # Combine test data features
    test_features0 = np.column_stack((avg_brightness_test0, std_brightness_test0))
    test_features1 = np.column_stack((avg_brightness_test1, std_brightness_test1))
    
    # Predict labels
    predictions0 = predict_class(test_features0, 
                                 (mean_f1_0, mean_f1_1), 
                                 (var_f1_0, var_f1_1), 
                                 (mean_f2_0, mean_f2_1), 
                                 (var_f2_0, var_f2_1))
    predictions1 = predict_class(test_features1, 
                                 (mean_f1_0, mean_f1_1), 
                                 (var_f1_0, var_f1_1), 
                                 (mean_f2_0, mean_f2_1), 
                                 (var_f2_0, var_f2_1))
    print("Predictions for Digit 0 Test Set:")
    print(predictions0)
    print("Predictions for Digit 1 Test Set:")
    print(predictions1)
    # Save predictions for later accuracy calculation
    np.save('predictions0.npy', predictions0)
    np.save('predictions1.npy', predictions1)

if __name__ == '__main__':
    main()
```

```python
### TEST FUNCTION: test_question4
import numpy as np
import scipy.io
from sklearn.metrics import accuracy_score

def load_test_labels():
    # Load the true labels for test data from MATLAB files
    testfile0 = scipy.io.loadmat('digit0_testset.mat')
    testfile1 = scipy.io.loadmat('digit1_testset.mat')
    
    test_labels0 = np.zeros(testfile0['target_img'].shape[0])
    test_labels1 = np.ones(testfile1['target_img'].shape[0])
    return test_labels0, test_labels1

def load_naive_bayes_params():
    # Load the Naive Bayes parameters
    params = np.load('naive_bayes_params.npz')
    return (params['mean_f1_0'], params['var_f1_0'], params['mean_f2_0'], params['var_f2_0'],
            params['mean_f1_1'], params['var_f1_1'], params['mean_f2_1'], params['var_f2_1'])

def main():
    myID = '7302'
    
    # Load the saved predictions
    predictions0 = np.load('predictions0.npy')
    predictions1 = np.load('predictions1.npy')
    
    # Load the true labels
    test_labels0, test_labels1 = load_test_labels()
    
    # Load Naive Bayes parameters
    (mean_f1_0, var_f1_0, mean_f2_0, var_f2_0,
     mean_f1_1, var_f1_1, mean_f2_1, var_f2_1) = load_naive_bayes_params()
    
    # Calculate accuracy
    accuracy0 = accuracy_score(test_labels0, predictions0)
    accuracy1 = accuracy_score(test_labels1, predictions1)
    
    # Print final output with headings
    output = [
        f"ID: {myID}",
        f"Mean of Feature 1 for Digit 0: {mean_f1_0}",
        f"Variance of Feature 1 for Digit 0: {var_f1_0}",
        f"Mean of Feature 2 for Digit 0: {mean_f2_0}",
        f"Variance of Feature 2 for Digit 0: {var_f2_0}",
        f"Mean of Feature 1 for Digit 1: {mean_f1_1}",
        f"Variance of Feature 1 for Digit 1: {var_f1_1}",
        f"Mean of Feature 2 for Digit 1: {mean_f2_1}",
        f"Variance of Feature 2 for Digit 1: {var_f2_1}",
        f"Accuracy for Digit 0: {accuracy0}",
        f"Accuracy for Digit 1: {accuracy1}"
    ]
    #Save results in the file
    with open('final_output1.txt', 'w') as file:
        file.write('Final Output:\n')
        file.write('\n'.join(output))
    print("Final Output:")
    for line in output:
        print(line)

if __name__ == '__main__':
    main()
```
